```python
df['10board'] = df['10board'].replace(['0'], df['10board'].mode())
df['12board'] = df['12board'].replace(['0'], df['12board'].mode())
```

```python
df = df.drop(['ID', 'CollegeID', 'CollegeCityID'], axis = 1)
```

```python
def get_age(DOB):
  dob = datetime.strptime(DOB, "%Y-%m-%d").date()
  today = date.today()
  return today.year - dob.year
```

```python
df['DOB'] = df['DOB'].apply(get_age)
```
