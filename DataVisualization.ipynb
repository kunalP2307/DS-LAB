```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df['sex'].value_counts().plot(kind = 'bar')
```

```python
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
df['deck'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
plt.hist(df['age'], bins=5 )
```

```python
sns.distplot(df['age'])
```

```python
sns.boxplot(df['age'])
```

```python
sns.boxplot(df['fare'])
```

```python
sns.scatterplot(x=df['age'], y=df['survived'])
```

```python
sns.scatterplot(x=df['pclass'], y=df['age'])
```

```python
sns.scatterplot(x=df['survived'], y=df['fare'])
```

```python
sns.barplot(x=df['pclass'],y=df['age'],hue=df['sex'])
```

```python
sns.boxplot(data=df, x='sex',y='age',hue='survived')
```

```python
sns.distplot(df[df['survived']==0]['age'],hist=False, label='Not urvived')
sns.distplot(df[df['survived']==1]['age'],hist=False, label='Survived')
plt.legend()
     
```

```python
sns.heatmap(pd.crosstab(df['class'],df['survived']))
```

```python
sns.pairplot(data=df,hue='survived')
```

```python
sns.lineplot(data=df, x='fare', y='age')
```
